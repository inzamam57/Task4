```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
df = pd.read_csv("/content/Adult_Income_Dataset.csv")
df.head()
```

```python
categorical_features = df.select_dtypes(include=["object"]).columns
numerical_features = df.select_dtypes(include=["int64", "float64"]).columns

print("Categorical Features:\n", categorical_features)
print("\nNumerical Features:\n", numerical_features)
```

```python
label_encoder = LabelEncoder()
df["education_encoded"] = label_encoder.fit_transform(df["education"])
```

```python
df_encoded = pd.get_dummies(
    df,
    columns=[
        "workclass",
        "marital.status",
        "occupation",
        "relationship",
        "race",
        "sex",
        "native.country",
        "income"
    ],
    drop_first=True
)
```

```python
scaler = StandardScaler()
df_encoded[numerical_features] = scaler.fit_transform(
    df_encoded[numerical_features]
)
```

```python
print("Before Scaling:\n", df[numerical_features].describe())
print("\nAfter Scaling:\n", df_encoded[numerical_features].describe())
```

```python
df_encoded.to_csv("/content/Adult_Income_Processed.csv", index=False)
```
